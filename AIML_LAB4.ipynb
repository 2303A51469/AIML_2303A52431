```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt


# Load the dataframe.
df = pd.read_csv("/content/sample_data/california_housing_train.csv")

# Check for the correct column name
print(df.columns)

# Now drop the column with the correct name
X = df.drop('median_house_value', axis=1) #The column name is actually median_house_value, not price

y = df['median_house_value'] # Use the correct column name here as well.


# Add a constant to the model (for the intercept)
X_const = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X_const).fit()
```

1.Calculate the standard error and t-statistic for the coe cients.

```python
summary = model.summary()
print(summary)

# Extract coefficients and their standard errors
coefficients = model.params
standard_errors = model.bse
t_statistics = coefficients / standard_errors
```

2.Sort all the coe cients based on the cdf. Generate the scatter plots for the other features considering price as dependent variable.

```python
cdf_sorted = coefficients.sort_values(ascending=False)
print("\nCoefficients sorted based on CDF:")
print(cdf_sorted)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
np.random.seed(0)
X = pd.DataFrame({
    'feature1': np.random.rand(100),
    'feature2': np.random.rand(100),
    'feature3': np.random.rand(100)
})
y = 3 * X['feature1'] + 2 * X['feature2'] + 4 * X['feature3'] + np.random.randn(100)
X_values = X.values  # Convert to NumPy array
y_values = y.values
model = LinearRegression()
model.fit(X_values, y_values)
y_pred = model.predict(X_values)
r2 = r2_score(y_values, y_pred)
print("R Score:", r2)
```

5.Generate the evaluation regression error metrics- MAE, SSE, RMSE ,R2 using metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
np.random.seed(0)
y = np.random.rand(100) * 100  # True values
y_pred = y + (np.random.rand(100) * 10 - 5)
mae = mean_absolute_error(y, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
sse = np.sum((y - y_pred) ** 2)
print(f"Sum of Squared Errors (SSE): {sse}")
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse}")
r_squared = r2_score(y, y_pred)
print(f"R: {r_squared}")
```

#2. Compute the MinMax value between Observed Price and Expected Price for the US Housing Data
1. Write the python code to compute MinMax value of a Feature within Housing data.
 We compute the MinMax value using the equation.

  L_minmax = (L_minmax-min(L_minmax))/
   (max(L_minmax)-min(L_minmax))
2. Normalize the data and Print the MinMax value, plot the distribution of feature

```python
import pandas as pd
data = {
      'price': [200000, 450000, 300000, 500000, 250000, 400000],
                'area': [1500, 2000, 1600, 2500, 1800, 2200]
}
df = pd.DataFrame(data)
def min_max_scaling(feature):
      min_value = feature.min()
      max_value = feature.max()
      return (feature - min_value) / (max_value - min_value)
df['price_minmax'] = min_max_scaling(df['price'])
print(df[['price', 'price_minmax']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
      'price': [200000, 450000, 300000, 500000, 250000, 400000],
          'area': [1500, 2000, 1600, 2500, 1800, 2200]
}
df = pd.DataFrame(data)
def min_max_scaling(feature):
      min_value = feature.min()
      max_value = feature.max()
      normalized_feature = (feature - min_value) / (max_value - min_value)
      return normalized_feature, min_value, max_value
df['price_minmax'], min_price, max_price = min_max_scaling(df['price'])
print("Original Price Min:", min_price)
print("Original Price Max:", max_price)
print("Normalized Price Min:", df['price_minmax'].min())
print("Normalized Price Max:", df['price_minmax'].max())
plt.figure(figsize=(8, 5))
plt.hist(df['price_minmax'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.title('Distribution of Normalized Price')
plt.show()
```
