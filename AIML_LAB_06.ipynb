Part 2 Model training and Fit the data to Model
1. Split the data generated from list created as X, Y is distributed using train test split
   function as X trainY trainX testY test
2. Apply the KNNClassi er modelof sklearn.neighbors import KNeighborsClassifier
     package
3. Fit the data to the Classier Model using fit

```python
# Import the required library
from sklearn.model_selection import train_test_split

# Assuming `X` contains the features and `Y` contains the target/labels
# Replace the definitions of X and Y as per your dataset if needed

# Assuming the first column is the target variable
# Update 'TARGET_COLUMN_NAME' with the actual name if different
TARGET_COLUMN_NAME = data.columns[0]

# Extract features (X) and target (Y)
X = data.drop(columns=[TARGET_COLUMN_NAME])
Y = data[TARGET_COLUMN_NAME]

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Print the shapes of the splits for verification
print("Training and testing data shapes:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"Y_test shape: {Y_test.shape}")
```

```python
# Import the required library
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize the KNN Classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)  # Set n_neighbors to 5 (default)

# Fit the model to the training data
knn_classifier.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = knn_classifier.predict(X_test)

# Evaluate the model
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred))

# Print the accuracy score
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy of the KNN Classifier: {accuracy:.2f}")
```

```python
# Import the required library
from sklearn.neighbors import KNeighborsClassifier

# Initialize the KNN Classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)  # Set n_neighbors to 5 (default)

# Fit the classifier model to the training data
knn_classifier.fit(X_train, Y_train)

# Print a confirmation message
print("KNN Classifier model has been successfully fitted to the training data.")
```

Part 3 Evaluate the Classification Quality
1. Generate the confusion matrix to estimate the correction among features
2. Generate the classification report using classification
    report

```python
# Import required libraries
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the model has already been trained with knn_classifier.fit(X_train, Y_train)

# 1. Generate the confusion matrix
Y_pred = knn_classifier.predict(X_test)  # Make predictions on the test data
conf_matrix = confusion_matrix(Y_test, Y_pred)

# Visualize the confusion matrix using seaborn's heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=knn_classifier.classes_, yticklabels=knn_classifier.classes_)
plt.title("Confusion Matrix")
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# 2. Generate the classification report
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred))

# Optionally, print accuracy score
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy of the KNN Classifier: {accuracy:.2f}")
```
